```python
train_dataset['HomePlanet'] = train_dataset['HomePlanet'].replace({'Mars':1, 'Earth':2, 'Europa':3})
train_dataset['Destination'] = train_dataset['Destination'].replace({'TRAPPIST-1e':1, 'PSO J318.5-22':2, '55 Cancri e':3})
train_dataset['VIP'] = train_dataset['VIP'].replace({False: 0, True: 1})
```

```python
categories = ['HomePlanet','Destination', 'VIP','Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
for category in categories:
  test_dataset[category].replace([np.nan], test_dataset[category].mean(), inplace=True)
```

FITTING THE DATASET

```python
columns_to_drop = ['Cabin', 'Name', 'Transported_Style']
X_train = train_dataset.drop(columns_to_drop, axis=1)
X_test = test_dataset.drop(columns_to_drop, axis = 1)
```

```python
Y_train = train_dataset["Transported_Style"]
Y_test = test_dataset["Transported_Style"]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

OBSERVE THE DATASET BY USING THE GRAPH

```python
from sklearn.neighbors import KNeighborsClassifier
training_accuracy = []
test_accuracy = []
for n_neighbors in range(1, 10):
    kNN = KNeighborsClassifier(n_neighbors = n_neighbors)

    kNN.fit(X_train, Y_train)
    training_accuracy.append(kNN.score(X_train, Y_train))
    test_accuracy.append(kNN.score(X_test, Y_test))

plt.plot(range(1,10), test_accuracy, label = "test_accuracy")
plt.plot(range(1,10), training_accuracy, label = "training_accuracy")
plt.xlabel("n_neighbors")
plt.ylabel("Accuracy of the data")
plt.legend()
plt.show()
```
