IMPORT THE LIBRARY

```python
import pandas as pd
df1 = pd.read_csv('Consumer_Test.csv')
#test dataset is df1
df2=pd.read_csv('Consumer_Train.csv')
# train dataset is df2
df1.head(10)
```

LODING THE TRAIN DATASET

```python
print('consumer train dataset')
df2.head(10)
```

PREPROCESS THE DATA

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('punkt')
#here nltk.download('pulkt') use for the tokenize because it give the error
nltk.download('stopwords')
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(token.lower()) for token in tokens if token.lower() not in stop_words]
    preprocessed_text = ' '.join(tokens)
    return preprocessed_text
df2['preprocessed_text'] = df2['Consumer complaint narrative'].apply(preprocess_text)
df1['preprocessed_text'] = df1['Consumer complaint narrative'].apply(preprocess_text)
print("Preprocessed Training Data:")
df2.head()
```

PREPROCESSED TEST DATA

```python
print("Preprocessed Test Data:")
df1.head()
```

EXTRACTED THE DATASET

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(df2['preprocessed_text'])
y_train = df2['Product']

X_test = vectorizer.transform(df1['preprocessed_text'])
y_test = df1['Product']

print("training data shape")
print(X_train.shape)

print("\ntest data shape")
print(X_test.shape)
```

FOR CLASSIFIER USING THE SKLEARN.SVM FUNCTION

```python
from sklearn.svm import SVC

classifier = SVC(kernel='linear')
classifier.fit(X_train, y_train)

print("Classifier Training Complete.")
```

PREDICTION THE DATASET AND F1 SCORE

```python
from sklearn.metrics import classification_report, f1_score

y_pred = classifier.predict(X_test)
f1 = f1_score(y_test, y_pred, average='weighted')
classification_report = classification_report(y_test, y_pred)

print(f"F1 Score: {f1}")
print("Classification Report:")
print(classification_report)
```
